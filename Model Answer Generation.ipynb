```python
import gensim
import pandas as pd
from gensim.models import Word2Vec
model_name = "10epoch4window.model"
```

```python
df = pd.read_csv("nyt.csv")
```

```python
df.head()
```

```python
df["data"] = df.clue.apply(gensim.utils.simple_preprocess)
```

```python
df.head()
```

```python
###Load in correct model
model = Word2Vec.load(model_name)
```

```python
print(model)
```

```python
def sumList(listy):
    total = 0
    count = len(listy)
    for i in listy:
        try:
            total += model.wv[i]
        except:
            count = count - 1
    if count == 0:
        return 0
    return total/count
```

```python
sumList(df["data"][2])
```

```python
df["avgVec"] = df["data"].apply(sumList)
```

```python
df.head()
```

```python
def top1000(vec):
    return model.wv.most_similar(vec, topn = 1000)
```

```python
df["top_1000"] = df["avgVec"].apply(top1000)
```

```python
df.head()
```

```python
def specificGuess(length, tuples):
    guesses = []
    for i in tuples:
        #print(len(i))
        if len(i[0]) == length:
            if len(guesses) < 50:
                guesses.append(i)
            else:
                return guesses
    return guesses
```

```python
def makeGuesses(df):
    listOfGuesses = []
    for i in range(0, len(df)):
        #print(df["length"][i])
        listOfGuesses.append(specificGuess(df["length"][i], df["top_1000"][i]))
    return listOfGuesses
        
```

```python
guesses = makeGuesses(df)
```

```python
finalList = []
for i in guesses:
    listOfFirst = []
    for j in i:
        listOfFirst.append(j[0])
    finalList.append(listOfFirst)
df["final_list"] = finalList
    
answers = []
for i in df["final_list"]:
    sentence = " ".join(i)
    sentence = sentence.upper()
    answers.append(sentence)
```

```python
df["answer"] = answers
```

```python
df.head()
```

```python
finalResults = df["answer"]
model_name_csv = model_name + ".csv"
finalResults.to_csv(model_name_csv, index_label="index")
```
